### Third Script - Perform division based on CSVs

This script is used after compiling frequency of overall counts and frequency counts of specific search terms.

```python
# Libraries for parsing data
from lxml import etree
from bs4 import BeautifulSoup
import pandas as pd
import os

# Library for plotting data
import matplotlib.pyplot as plt
```

```python
search_df = pd.read_csv('/home/ec2-user/SageMaker/data/2023-06-18-export_newsample/2023-06-22-newsample_3_rawcount.csv')
overall_df = pd.read_csv('/home/ec2-user/SageMaker/data/2022-10-06-test_export_files_1/ALL/2022-11-12-ALLNYT.csv')
```

```python
search_df['yearmonth'] = search_df[['year','month']].apply(lambda x: '-'.join(x.values.astype(str)), axis='columns')
search_df['yearmonth'] = pd.to_datetime(search_df['yearmonth'])
search_df = search_df.set_index(search_df['yearmonth'])
search_df = search_df.drop(['year','month','yearmonth'], axis='columns')
search_df = search_df.sort_index()
search_df = search_df[(search_df.index >= '1855-01-01') & (search_df.index <= '2018-12-31')]
```

```python
overall_df['yearmonth'] = overall_df[['year','month']].apply(lambda x: '-'.join(x.values.astype(str)), axis='columns')
overall_df['yearmonth'] = pd.to_datetime(overall_df['yearmonth'])
overall_df = overall_df.set_index(overall_df['yearmonth'])
overall_df = overall_df.drop(['year','month','yearmonth', 'Unnamed: 0'], axis='columns')
overall_df = overall_df.sort_index()
```

```python
division_df = search_df.div(overall_df)
```
